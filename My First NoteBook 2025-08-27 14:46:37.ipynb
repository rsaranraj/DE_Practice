```python
#1: How do you create a DataFrame in Spark from a collection of data?
from pyspark.sql import SparkSession
#initilaise spark session
spark=SparkSession.builder.appName("createDataFrame").getOrCreate()
#sample data
data=[("john",25),("Doe",30),("Jane",28)]
#create Dataframe
df=spark.createDataFrame(data,schema=["name","age"])
#display dataframe
df.show()
```

```python
#1: How do you create a DataFrame in Spark from a collection of data?
display(df.select("name","age"))
```

```python
#How do you filter rows in a DataFrame based on a condition?
from pyspark.sql.functions import col
df.filter(col("age")>25).show()
```

```python
#How do you group by a column and perform an aggregation in Spark DataFrame? 
emp_df=spark.read.option("header","True")\
                 .option("inferSchema","True")\
                  .csv("/Volumes/hr_catalog/hr_schema/hr_volume/hr_directory/employees.csv") 
```

```python
spark.sql("select department_id,count(department_id) from hr_catalog.hr_schema.employees group by department_id").display()
```

```python
display(emp_df.groupBy("DEPARTMENT_ID").count())
```
